```python
import cv2
import numpy as np
import sys
from os.path import exists
from IPython.display import Image
```

```python
# Image qu'on recherche (objet en l'occurence)
template = cv2.imread('./questItem/peacekeeper/propane.png')

# Image dans laquelle on recherche (scav1.png)
img_rgb = cv2.imread("C:/Users/mohamed/Desktop/projet_perso/tarkovSorter/img/scav1.png")
```

```python
# Taille de la cellule qu'on cherche.
w, h = template.shape[:-1]
```

```python
# Recherche de l'objet dans l'image.
res = cv2.matchTemplate(img_rgb, template, cv2.TM_CCOEFF_NORMED)
```

```python
# On garde les cellules qui ont une valeur superieure au seuil.
threshold = .70
loc = np.where(res >= threshold)

trows,tcols = template.shape[:2]
```

```python
# Nombre de correspondance
print((list(zip(*loc[::-1]))))
```

```python
# Elimination des doublons
find = list(zip(*loc[::-1]))[0]
print(find)
for pt in zip(*loc[::-1]):
    pass
```

```python

# Création des rectancles sur les correspondances.
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + tcols, pt[1] + trows), (0, 0, 255), 2)
```

```python
# Enregistrement des résultats
cv2.imwrite('dump/result.png', img_rgb)
cv2.waitKey()
```
